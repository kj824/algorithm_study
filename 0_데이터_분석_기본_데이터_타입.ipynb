- 역원: $a$의 역원은 $-a$
- 교환법칙: $a+b=b+a$

## **2-2-2. float (실수)**



### **실수 (Real number, **ℝ**)의 정의**

### **직관적 정의**

- 실수는 **수직선 위의 모든 점에 대응되는 수**의 집합이다.

- 즉 “연속적인 크기(길이, 시간, 온도 등)”를 표현하기 위해 등장하는 수 체계다.

<br>

### **공리적 관점**

- 실수 집합 $(\mathbb{R})$는 다음을 만족하는 구조로 정의할 수 있다.

$$(\mathbb{R}, +, \cdot, \le)$$

가

1. **순서체(ordered field)** 이고
2. **완비성(least upper bound property)** 을 갖는다.

- 여기서 “완비성”은

  - $\mathbb{R}$의 공집합이 아닌 부분집합 $S\subseteq\mathbb{R}$이 위로 유계(upper bounded)이면, S는 **최소상계(supremum)** 를 가진다.

- 이 성질 때문에 $\mathbb{R}$에서는 “구멍이 없다”는 의미의 연속성이 성립.

<BR>

### 3) **포함 관계(정리)**

$$\mathbb{Z}\subset\mathbb{Q}\subset\mathbb{R}$$

-$\mathbb{Z}$: 정수
- $\mathbb{Q}$: 유리수(분수로 표현 가능)
- $\mathbb{R}$: 유리수 + 무리수(예: $\sqrt2,\pi$) 전부 포함

<BR>

### 4) **영어 용어(왜 Real인가)**

- **real number**라는 표현은 “현실에 존재한다” 같은 뜻에서 시작됐다기보다, 역사적으로 **“허수(imaginary)”와 대비되는 이름**으로 굳어진 용어.
  - 17세기 이후 복소수 체계가 정리되면서, $\sqrt{-1}$ 같은 수를 **imaginary**라고 부르고, 그에 대비해 기존의 수(복소수에서 허수부가 0인 부분)를 **real**이라 부르게 됐다.
- 즉 “real = not imaginary(허수가 아닌)”이라는 **대비 용법**.
    



#### **쓸없수지 : 최소상계(least upper bound)란?**
- “어떤 집합을 위에서 덮는 상계들 중 가장 작은 것”.

<br>

#### 1) **상계(upper bound)부터**

실수 집합의 부분집합 $(S \subset \mathbb{R})$에 대해,

* 어떤 수 (u)가 **상계**라는 건
  $$
  \forall x \in S,; x \le u
  $$
  가 성립하는 것.
  즉, $u$는 $S$의 모든 원소보다 **크거나 같은** “천장”.

예) (S={1,2,3})이면 3, 4, 100은 다 상계.

<br>
<br>

#### 2) **최소상계(least upper bound) = 슈프리멈(supremum)**

- 상계는 보통 여러 개가 있음.
- 그중에서 **가장 작은 상계**를 **최소상계**라고 하고, 이를 아래같이 씀
$$
\sup S
$$


- **정의**
  - 어떤 수 $s$가 $S$의 슈프리멈이라는 건 두 조건을 동시에 만족하는 것.

1. $s$는 상계:
   $$
   \forall x\in S,; x\le s
   $$
2. $s$보다 더 작은 상계는 없다 $“가장 작다”$:
   $$
   \forall u ;(\text{u가 상계}) \Rightarrow s \le u
   $$

<br>

- 또는 많이 쓰는 동치 정의($ε$ 정의):

* $s$는 상계이고,
* 임의의 $\varepsilon>0$에 대해 $S$ 안에 $s-\varepsilon$보다 큰 원소가 존재:
  $$
  \forall \varepsilon>0,; \exists x\in S \text { such that } s-\varepsilon < x \le s
  $$
  즉, $S$의 원소들이 $s$에 **끝없이 가까이** 접근한다는 뜻.

<br>
<br>

#### 3) **최대값(max)와 차이**

* **최대값** $ \max S $: 집합 안에 실제로 존재하는 “가장 큰 원소”
  $$
  m\in S \text{ 이고 } \forall x\in S,; x\le m
  $$
* **슈프리멈** $ \sup S $: 집합 밖에 있을 수도 있는 “가장 낮은 천장”

관계:

* $\max S$가 존재하면 항상 $\sup S = \max S$
* 하지만 $\sup S$는 존재해도 $\max S$가 없을 수 있음

### **6) 문자 성질 검사 메서드 (is~ 계열)**

### **6-1. `isdigit()`**

- 전부 “숫자 문자”면 True 나옴
- 단, `1`, `3.14` 같은 건 False 나옴(마이너스/점 때문에)
- 전각 숫자(예: `１２３`) 같은 것도 True 될 수 있음

```python
print("123".isdigit())     # True
print("-123".isdigit())    # False
print("3.14".isdigit())    # False
print("12a".isdigit())     # False
```

### **6-2. `isnumeric()` / `isdecimal()`**

- `isdecimal()` 가장 빡셈(10진 숫자만)
- `isdigit()` 그 다음
- `isnumeric()` 제일 넓음(분수/특수 숫자 문자까지 포함 가능)

```python
print("123".isdecimal())   # True
print("123".isdigit())     # True
print("123".isnumeric())   # True
```

#### **6-3. `isalpha()` / `isalnum()`**
- `isalpha()` 전부 글자면 True(공백/숫자 있으면 False)
- `isalnum()` 글자+숫자만 있으면 True(특수문자 있으면 False)

```python
print("abc".isalpha())
print("abc1".isalpha())

print("abc1".isalnum())
print("abc_1".isalnum())
```

#### **1) 비어있다 / 비어있지 않다**
- 컨테이너(list/tuple/dict/set/str)는 **비어있으면 False**, 비어있지 않으면 True로 취급.

```python
items = []
print(not items)
```

```python
items = [1, 2]
print(not items)
```

```python
s = ""
print(not s)
```

#### **2) None(값 없음) 체크**
- `None`은 “값이 없다”는 표시라서, 보통 이렇게 확인해.

```python
x = None
print(x is None)
```

```python
print(not (x is None))
```

#### **3) “공백만 있는 문자열”도 ‘없다’로 보고 싶을 때**

- 빈 문자열이 아닌 `"   "` 같은 공백 문자열도 “없음” 처리하고 싶으면 `strip()`을 같이 씀.

```python
s = "   "
print(not s)
print(not s.strip())
```

#### **4) 포함/미포함을 not으로**

`in` 결과(True/False)에 `not`을 붙이면 “없다”가 됨.

```python
nums = [1, 2, 3]
print(not (5 in nums))
print(not (2 in nums))
```

```python
di = {'a':1,'b':2}

for i in di:
  print(i)
```

```python
d = {"a": 1}
print( (1 in d.values()))
```

#### **6) 조건 여러 개를 not으로 묶기**

- 자주 쓰는 논리 변형.

* `not (A and B)` 는 `not A or not B` 와 같은 의미
* `not (A or B)` 는 `not A and not B` 와 같은 의미

```python
A = True
B = False

print(not (A and B))
print((not A) or (not B))
```
