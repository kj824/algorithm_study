### ---문자 성질 검사 메서드 (is~ 계열)---

- 문자열이 “숫자냐/문자냐/공백이냐” 같은 거 빠르게 체크할 때 씀

#### **(1) `isdigit()`**

- 전부 “숫자 문자”면 True 나옴
- 단, `1`, `3.14` 같은 건 False 나옴(마이너스/점 때문에)
- 전각 숫자(예: `１２３`) 같은 것도 True 될 수 있음

```python
print("123".isdigit())     # True
print("-123".isdigit())    # False
print("3.14".isdigit())    # False
print("12a".isdigit())     # False
```

```python
from os import replace
# 정수를 판별하는 함수

def Z_판별기(정수):
    if  정수.replace('-', '').isdigit():
        return True
    else:
        return False

'123a'
'222'
'-0.5'
Z_판별기('222')

#.(점)은 true로 안됨
```

#### **(2) `isnumeric()` / `isdecimal()`**

- `isdecimal()` 가장 빡셈(10진 숫자만)
- `isdigit()` 그 다음
- `isnumeric()` 제일 넓음(분수/특수 숫자 문자까지 포함 가능)

```python
print("123".isdecimal())   # True
print("123".isdigit())     # True
print("123".isnumeric())   # True
```

#### **(3) `isalpha()` / `isalnum()`**
- `isalpha()` 전부 글자면 True(공백/숫자 있으면 False)
- `isalnum()` 글자+숫자만 있으면 True(특수문자 있으면 False)

```python
print("abc".isalpha())
print("abc1".isalpha())

print("abc1".isalnum())
print("abc_1".isalnum())
```

#### **(4) `isspace()`**

- 공백(스페이스/탭/개행)만 있으면 True

```python
print("   ".isspace())     # True
print("\n\t".isspace())    # True
print(" a ".isspace())     # False

# print('\n\t')
```

## bool (불리언: True/False)



### **1) bool 기본**

- `bool`은 **True / False** 두 값만 갖는 타입이고, 비교/조건문의 결과로 가장 많이 등장.

```python
print(10 > 3)   # True
print(10 == 3)  # False
```

### **2) 논리 연산 (and, or, not)**

```python
a, b = True, False

print(a and b)  # False (둘 다 True여야 True)
print(a or b)   # True  (하나라도 True면 True)
print(not a)    # False (뒤집기)
```

### **3) Truthy / Falsy (자동으로 bool로 해석되는 규칙)**

파이썬은 조건문에서 다음 값들을 **False처럼** 취급:

- `0`, `0.0`
- `""` (빈 문자열)
- `[]`, `{}`, `set()` (빈 컨테이너)
- `None`

```python
print(bool(0))
```

```python
bool("")
#빈 문자열만 False, 빈 문자열이 아니라면 True
```

```python
bool([])
```

```python
bool(None)
```

```python
print(bool("0"), bool([0]))                     # True True
```

```python
print(bool(0))     # False
print(bool(""))    # False
print(bool([]))    # False
print(bool(None))  # False
print(bool("0"), bool([0])) #True True
```

### **4) `all()` 과 `any()`**

- 둘 다 “여러 조건을 한 번에” 판단할 때 쓰는 함수.



#### **4-1) `all(iterable)`**

- **모든 값이 True(Truthy)** 이면 True
- 하나라도 False(Falsy)면 False

```python
print(all([True, True, True]))
```

```python
print(all([True, False, True]))
```

```python
print(all([1, "hi", [0]]))
```

```python
print(all([1, "", 3]))
```

#### **4-2) `any(iterable)`**

- **하나라도 True(Truthy)** 이면 True
- 전부 False(Falsy)면 False


```python
print(any([False, False, True]))
```

```python
print(any([0, "", None]))
```

```python
nums = [0, 0, 5, 0]
```

```python
print(any(n != 0 for n in nums))
```

```python
vals = [1, None, 3]
print(any(v is None for v in vals))  # True
```

### **5) `all/any` 사용할 때 중요한 포인트**

#### **비어있는 경우(빈 리스트)의 결과**

- `all([])` 는 **True**
- `any([])` 는 **False**


```python
print(all([]))  # True
print(any([]))  # False

#외우기....
```

#### **6) 조건 여러 개를 not으로 묶기**

- 자주 쓰는 논리 변형.

* `not (A and B)` 는 `not A or not B` 와 같은 의미
* `not (A or B)` 는 `not A and not B` 와 같은 의미

```python
A = True
B = False

print(not (A and B))
print((not A) or (not B))
```
